```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Most Subscribed YouTube Channels_exported.csv')
```

```python
df.head()
```

```python
x=df['Brand channel']
y=df['Subscribers (millions)']
```

```python
plt.scatter(x,y)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(data=df, x='Category', palette='Set2')
plt.xticks(rotation=40)
plt.title('Youtube Best Category Chart')
```

Above graph shows that the best catogory of Youtube channel is Music

```python
plt.figure(figsize=(13,6))
top_countries=df['Country'].value_counts().head(8).index
top_countries
df_top_countries=df[df['Country'].isin(top_countries)]
sns.countplot(data=df_top_countries,x='Country', palette='Set1')
```

Graph indicates that the  Highest number of the subscriber are from India

```python
top_channel = df.sort_values(by='Subscribers (millions)',ascending=False).head(10)
color=['Red','Green','orange','grey','blue','yellow','gold','silver','pink','Black']
plt.figure(figsize=(13,6))
plt.bar(top_channel['Name'],top_channel['Subscribers (millions)'], color=color)
plt.xticks(rotation=90)
```
