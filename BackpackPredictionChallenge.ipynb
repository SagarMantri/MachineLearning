```python
from sklearn.model_selection import train_test_split
Target=df['Price']
x=df.drop(columns='Price')
y=Target


```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=41)
```

```python
from sklearn.pipeline import make_pipeline
from tqdm import tqdm
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder
from sklearn.metrics import mean_squared_error
```

```python
xgboost_params={'max_depth':8,
                'n_estimators':1701,
                'learning_rate':0.000678,
                'min_child_weight':0.275,
                'reg_alpha':0.527,
                'reg_lambda':13.91,
                'colsample_bylevel':0.977,
                'colsample_bytree':0.777,
                'colsample_bynode':0.8541
                }
lgd_params={'max_depth':17,
            'n_estimators':2771,
            'learning_rate':0.009,
            'min_child_weight':0.92,
            'reg_alpha':0.92,
            'reg_lambda':1.48,
            'colsample_bylevel':0.52,
            'colsample_bytree':0.99,
            'colsample_bynode':0.84}

algorithms=[LGBMRegressor(**lgd_params,verbose=-1,random_state=42),
            XGBRegressor(**xgboost_params,random_state=42)]

for algorithm in  tqdm(algorithms, desc='Starting'):
   model=make_pipeline(
      OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1),
      StandardScaler(),
      algorithm
  )
   alg_name=list(model.named_steps.keys())[-1]
#   alg_name = list(model.named_steps.keys())[-1]
   print(f"\nFitting the {alg_name}")
   model.fit(x_train,y_train)
   y_pred=model.predict(x_test)
   score=np.sqrt(mean_squared_error(y_test,y_pred))
   print(f"\nThe Score with the algorithm {alg_name} is {score}")
```
